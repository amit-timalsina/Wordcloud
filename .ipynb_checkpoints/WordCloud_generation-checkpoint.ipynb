```python
# Import packages
import wikipedia
import re# Specify the title of the Wikipedia page
from wordcloud import WordCloud, STOPWORDS #Generate word cloud
wiki = wikipedia.page('Data analysis')# Extract the plain text content of the page
text = wiki.content# Clean text
text = re.sub(r'==.*?==+', '', text)
text = text.replace('\n', '')
print(text[:500])
```

```python
import matplotlib.pyplot as plt
# %matplotlib inline# Define a function to plot word cloud
def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(40, 30))
    # Display image
    plt.imshow(wordcloud, interpolation='bilinear') 
    # No axis details
    plt.axis("off");
```

```python
wordcloud = WordCloud(width = 600, height = 400, random_state=1, background_color='green', colormap='Set2', collocations=False, stopwords = STOPWORDS).generate(text)# Plot
plot_cloud(wordcloud)
plt.show()
```

```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import wikipedia
import re# Specify the title of the Wikipedia page
from wordcloud import WordCloud, STOPWORDS #Generate word cloud
wiki = wikipedia.page('Pakistan')# Extract the plain text content of the page
pak_text = wiki.content# Clean text
pak_text = re.sub(r'==.*?==+', '', pak_text)
pak_text = pak_text.replace('\n', '')
pak_text[:500]
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["country", "countries"])
```

```python
# Generate a word cloud image
mask = np.array(Image.open("1280px-Flag_of_Pakistan.svg.png"))
wordcloud_usa = WordCloud(height=600, width=800, stopwords=stopwords, background_color="white", mode="RGBA", max_words=1000, mask=mask).generate(pak_text)

# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[7,7])
plt.imshow(wordcloud_usa.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")

# store to file
# plt.savefig("img/us_wine.png", format="png")

plt.show()
```
